Gompertz curve
=============

Model
-----

$$f\left(t\right) = a\mathrm{e}^{-b\mathrm{e}^{-ct}}$$

where
- $a$ is an asymptote, since $\lim_{t \rightarrow \inf} a\mathrm{e}^{-b\mathrm{e}^{-ct}} = a\mathrm{e}^{0}=a$
- $b$ sets the displacement along the x-axis (translates the graph to the left or right).
- $c$ sets the growth rate (y scaling)

The rate of growth:
$$r=\frac{f'(t)}{f(t)}=bc\mathrm{e}^{-ct}$$



Cofficients
-------------

$$ z_i = \log\Biggl(2\frac{y_{i+1}-y_i}{y_{i+1}-y_i}\Biggr) = \log\left(r\right) $$
$$ \sum\left(pt + q - z_i\right)^2 \rightarrow \min$$

$$ \sum\left(pt + q - y\right) t = 0 $$
$$ \sum\left(pt + q - y\right) = 0 $$

$$ p\left<t^2\right> + q\left<t\right> = \left<yt\right>$$
$$ p\left<t\right> + q = \left<y\right>$$

$$ D = \left<t^2\right> - \left<t\right>\left<t\right> $$
$$ D_p = \left<yt\right> - \left<t\right>\left<y\right>$$
$$ D_q = \left<t^2\right>\left<y\right> - \left<yt\right>\left<t\right>$$

$$ p = \frac{D_p}{D}, q = \frac{D_q}{D} $$

$$ bc\exp\left(-ct\right) = \exp\left(pt\right)\exp\left(q\right) $$

$$ c = -p, b = -\frac{\exp\left(q\right)}{p}, a = \left<y\exp\left(b\exp\left(-ct\right)\right)\right> $$

```python
country = 'Spain'
cs, dt = get_country_data(country)

y = cs.to_numpy().astype(float)
```

```python
i0=33
inv = [[i0, i0+29], [i0+29, y.size]]

show_inv(y, inv)
```

```python
prm = fit_all_inv(y, inv)
show_model(y, prm)
pp[country] = prm
```
